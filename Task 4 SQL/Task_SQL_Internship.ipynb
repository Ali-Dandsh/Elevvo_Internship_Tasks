```python
date=pd.read_sql("""select date(invoicedate) as date_only from invoice """,conn) 
date
```

```python
date_months = pd.read_sql("""
SELECT 
    strftime('%m', InvoiceDate) AS Month,
    COUNT(i.InvoiceId) AS Num_Invoices,
    COUNT(DISTINCT i.CustomerId) AS Num_Customers,
    SUM(il.Quantity) AS Total_Quantity,
    SUM(IL.UNITPRICE*IL.QUANTITY) AS Total_Sales
FROM Invoice AS i
INNER JOIN InvoiceLine AS il 
    ON i.InvoiceId = il.InvoiceId
GROUP BY strftime('%m', InvoiceDate)
ORDER BY Month;
""", conn)

date_months
```

```python
date_year=pd.read_sql("""SELECT 
    strftime('%Y', InvoiceDate) AS year,sum(total) as Total_sales from invoice group by year """,conn) 
date_year
```

```python
time_of_playing=pd.read_sql('''SELECT 
    g.Name AS Genre,
    t.Name AS TrackName,
    t.Milliseconds / 1000.0 AS DurationSeconds,
    RANK() OVER (PARTITION BY g.Name ORDER BY t.Milliseconds DESC) AS RankByLength
FROM Track t
JOIN Genre g ON t.GenreId = g.GenreId
ORDER BY g.Name, RankByLength;''',conn)
time_of_playing
```

```python
customers_and_orders=pd.read_sql('''SELECT 
    c.Country,
    c.CustomerId,
    c.FirstName || ' ' || c.LastName AS CustomerName,
    COUNT(i.InvoiceId) AS TotalInvoices,
    ROW_NUMBER() OVER (PARTITION BY c.Country ORDER BY COUNT(i.InvoiceId) DESC) AS RowNum
FROM Customer c
JOIN Invoice i ON c.CustomerId = i.CustomerId
GROUP BY c.Country, c.CustomerId
ORDER BY c.Country, RowNum;'''
,conn)
customers_and_orders
```
